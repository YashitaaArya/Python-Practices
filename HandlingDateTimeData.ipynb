# Working with DATES

```python
# Converting to datetime datatype
date['date']=pd.to_datetime(date['date'])
```

```python
date.info()
```

1. Extract year

```python
date['date_year']=date['date'].dt.year
date.head()
```

2. Extract Month

```python
date['date_month_no']=date['date'].dt.month
date.head()
```

```python
date['date_month_name']=date['date'].dt.month_name()
date.head()
```

3. Extract Days

```python
date['date_day']=date['date'].dt.day
date.head()
```

```python
date['date_day_of_week']=date['date'].dt.dayofweek
date.head()
```

```python
date['date_dow_name']=date['date'].dt.day_name()
date.head()
```

```python
# to check if weekend?
date['date_isWeekend']=np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.head()
#
```

4. Extract week of the year

```python
date['date_weekOfYear']=date['date'].dt.isocalendar().week
date.head()
```

5. Extract Quarter

```python
date['date_quarter']=date['date'].dt.quarter
date.head()
```

6. Extract Semester

```python
date['date_semester']=np.where(date['date_quarter'].isin([1,2]),1, 2)
date.head()
```

7. Extract Time elapsed between 2 dates

```python
import datetime
today=datetime.datetime.today()
today
```

```python
today-date['date']
```

```python
(today-date['date']).dt.days
```

```python
# months passed
np.round((today-date['date'])/np.timedelta64(1,'D')/30,0)
```

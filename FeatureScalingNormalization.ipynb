```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/wine_data.csv', header=None, usecols=[0,1,2]) #taking only 3 columns
df.columns=['Class label', 'Alcohol', 'Malic Acid']
```

```python
df
```

```python
sns.kdeplot(df['Alcohol'])
```

```python
sns.kdeplot(df['Malic Acid'])
```

```python
color_dict={
    1:'red',
    2:'blue',
    3:'green'
}
```

```python
sns.scatterplot(x=df['Alcohol'], y=df['Malic Acid'], hue=df['Class label'], palette=color_dict)
```

```python
sns.scatterplot(x=df['Alcohol'], y=df['Malic Acid'], hue=df['Class label'].map(color_dict))
```

```python
from numpy.random import random
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('Class label', axis=1),
                                                    df['Class label'],
                                                    test_size=0.3,
                                                    random_state=0)
x_train.shape, x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

scaler.fit(x_train)
# fit to training data but transform both train and test
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled, columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled, columns=x_test.columns)
```

```python
np.round(x_train.describe(), 1)
```

```python
np.round(x_train_scaled.describe(), 1)
```

```python
fig, (ax1, ax2)=plt.subplots(ncols=2, figsize=(12,5))
ax1.scatter(x_train['Alcohol'], x_train['Malic Acid'], c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(x_train_scaled['Alcohol'], x_train_scaled['Malic Acid'], c=y_train)
ax2.set_title("After Scaling")
plt.show()
# data is squished into a unit rectangle
```

```python
fig, (ax1, ax2)=plt.subplots(ncols=2, figsize=(12,5))

# before scaling
ax1.set_title("Before Scaling")
sns.kdeplot(x_train['Alcohol'], ax=ax1)
sns.kdeplot(x_train['Malic Acid'], ax=ax1)

# after scaling
ax2.set_title("After Scaling")
sns.kdeplot(x_train_scaled['Alcohol'], ax=ax2)
sns.kdeplot(x_train_scaled['Malic Acid'], ax=ax2)
plt.show()
# kde plot takes inference
```

```python
fig, (ax1, ax2)=plt.subplots(ncols=2, figsize=(12,5))

# before scaling
ax1.set_title("Alcohol Distribution Before Scaling")
sns.kdeplot(x_train['Alcohol'], ax=ax1)

# after scaling
ax2.set_title("Alcohol Distribution After Standard Scaling")
sns.kdeplot(x_train_scaled['Alcohol'], ax=ax2)
plt.show()
```

```python
# same with malic acid
fig, (ax1, ax2)=plt.subplots(ncols=2, figsize=(12,5))

# before scaling
ax1.set_title("Malic Acid Distribution Before Scaling")
sns.kdeplot(x_train['Malic Acid'], ax=ax1)

# after scaling
ax2.set_title("Malic Acid Distribution After Standard Scaling")
sns.kdeplot(x_train_scaled['Malic Acid'], ax=ax2)
plt.show()
```

Categorical Scatterplots : Stripplots & Swarmplots

```python
# x-axis : Categorical data
# y-axis : Numerical Data
plt.style.use('fivethirtyeight')
tips=sns.load_dataset('tips')
tips.head()
```

```python
# stripplot
sns.catplot(x='day',y='tip', kind='strip', data=tips)
# jitter parameter : induces noise small which separates values overlapping
```

```python
sns.catplot(x='day',y='tip', kind='strip', jitter=0 ,data=tips)
```

```python
# swarmplot
sns.catplot(x='day',y='tip', kind='swarm', data=tips)
# width shows high distribution of data
```

```python
sns.catplot(x='day', y='tip', kind='swarm', hue='sex', data=data)
```

```python
sns.swarmplot(x='day', y='tip', data=tips)
```

Heatmaps

```python
# either create a pivot table
x=flights.pivot_table(index='year',columns='month',values='passengers', aggfunc='sum')
# sns.catplot(x='month',y='passengers',kind='point',data=flights)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d')
#
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='YlGnBu')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='summer')
```

```python
sns.clustermap(x, z_score=0, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True, row_cluster=False,metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True, col_cluster=False,metric='correlation')
```

Jointplots

```python
# used to do bivariate analysis on data
# used either as scatterplots, hexbinplots, kdeplots, regplot
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
# it also plots the individual histograms on both axes
# along bivariate, it gives univariate analysis also
```

```python
sns.jointplot(x='total_bill',y='tip',kind='hex',data=tips)
# 2D histogram
# denser bin signifies higher frequency than others
```

```python
sns.jointplot(x='total_bill',y='tip',kind='kde',data=tips)
# imagine it like an upward/top view of a mountain where centre is peak
# countour plot like
```

```python
sns.jointplot(x='total_bill',y='tip', kind='reg',data=tips)
# draws a linesegment
```

Distplots

```python
# used to do univariate analysis on continuous data
# combines : histogram, kernel density estimation (PDF), rugplot
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
# to fill certain missing values in age with mean of age
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
sns.distplot(titanic['age'])
# graph is a combination of 2 diff graphs
```

```python
sns.distplot(titanic['age'],kde=False)
# plotting only the histogram
# hence axes will also change
```

```python
sns.distplot(titanic['age'], bins=100, kde=False)
```

```python
# plotting only kde
sns.distplot(titanic['age'], hist=False)
```

```python
# rugplot
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
# rugplot interpretation : plotting the points side by side
```

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
```

```python
# comparing ages of people who survived and died
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
